## Two Options for Subplots

```python
#1 way of plotting
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#Filling the subplots with some data
ax1.plot(x,x**2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(butter_store_sales.keys(),butter_store_sales.values());
ax4.plot(np.random.normal(0,1,20));
```

```python
# 2nd way of plotting Subplots
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#Filling the subplots based on indexes
ax[0,0].plot(x,x**2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(butter_store_sales.keys(),butter_store_sales.values());
ax[1,1].plot(np.random.randn(100));
```

## Plotting data from dataframe

```python
#Reading CSV
car_sales=pd.read_csv("car-sales.csv")
car_sales
```

```python
#Creating Total Sales column to store the Cumulative Sales Data
#Remember that the Price column Contains String so we have to reformat the column then find the Cumulative Sum
car_sales["Price"]=car_sales["Price"].replace('[\\$\\,\\.]','',regex=True)
car_sales
```

```python
#Removing the Last 2 Digits from Price
car_sales["Price"]=car_sales["Price"].str[:-2]
car_sales
```

```python
#want to generate the sold date for the cars 
car_sales["Sale Date"]=pd.date_range('30-06-2026',periods=len(car_sales))
car_sales
```

```python
#Adding Total Sales that is Cumulative Price
car_sales["Total Sales"]=car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
#Plotting total Total Sales Vs Date Range
fig,ax=plt.subplots(figsize=(20,10))
ax.plot(car_sales["Sale Date"],car_sales["Total Sales"]);
```

```python
#Another Way of Plotting Directly from Pandas
car_sales.plot(x="Sale Date",y="Total Sales")
```

```python
#Plotting a bar graph for Make Vs Odometer 
car_sales.plot(x="Make",y="Odometer (KM)",kind="bar")
```

```python
#Reading Heart Disease Csv from github
heart_disease=pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/refs/heads/master/data/heart-disease.csv")
heart_disease[:10]
```

```python
# Different way of ploting using pyplot vs OO method 
# Lets extract the patient details whose age is < 50
patient_50=heart_disease[heart_disease["age"]>50]
patient_50.head()
```

```python
#plotting using pyplot
patient_50.plot(kind="scatter",
               x="age",
               y="chol",
               c="target");
```

```python
#Not that promising right ?
# Let Plot it using OO method so that we can do some customisation
```

```python
# Plotting using OO method mixed with pyplot
fig,ax=plt.subplots()
# ax.scatter(x=patient_50["age"],y=patient_50["chol"],c=patient_50["target"]);
patient_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax)
ax.set(title="Scatter Plot for age > 50",xlabel="Age",ylabel="Cholesterol");
```

```python
# OO method plot from scratch
fig, ax=plt.subplots()

scatter=ax.scatter(x=patient_50["age"],y=patient_50["chol"],c=patient_50["target"]);
ax.set(title="Scatter Plot for age > 50",xlabel="Age",ylabel="Cholesterol");

# Adding a legend
ax.legend(*scatter.legend_elements(),title="Target");

# Adding a horizontal Line
ax.axhline(patient_50["chol"].mean(),linestyle="--")
```

```python
fig, (ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

scatter=ax0.scatter(x=patient_50["age"],y=patient_50["chol"],c=patient_50["target"]);
ax0.set(title="Heart Disease  Cholesterol Vs Age",ylabel="Cholesterol");

# Adding a legend
ax0.legend(*scatter.legend_elements(),title="Target");

# Adding a horizontal Line
ax0.axhline(patient_50["chol"].mean(),linestyle="--")


#ploting for ax1:
scatter=ax1.scatter(x=patient_50["age"],y=patient_50["thalach"],c=patient_50["target"]);
ax1.set(title="Heart Disease  Heart Rate Vs Age",ylabel="Heart Rate",xlabel="Age");

# Adding a legend
ax1.legend(*scatter.legend_elements(),title="Target");

# Adding a horizontal Line
ax1.axhline(patient_50["thalach"].mean(),linestyle="--");

#Adding Title to the Figure
fig.suptitle("Heart Diseases Analysis",fontsize=16,fontweight="bold");
```
