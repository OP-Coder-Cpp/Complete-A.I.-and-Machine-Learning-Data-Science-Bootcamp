## Manipulating Data in Pandas

```python
#string operations converting to lowercase using the method lower():
car_sales["Make"].str.lower()
```

```python
#filling missing values
car_sales_missing=pd.read_csv("car-sales-missing-data.csv")
```

```python
car_sales_missing
```

```python
#we are going to fill the missing data in the dataframe NaN with mean value of that column , just a practice 
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
#but it will not update in the original dataframe we need to either assign the updated value or use inplace parameter. by Default inplace=False
```

```python
#inplace=True this thing got deprecated so we are going to use the assignment operator
```

```python
car_sales_missing["Odometer"]=car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales_missing
```

```python
#we can also delete the missing values 
temp=car_sales_missing.dropna()
temp
```
