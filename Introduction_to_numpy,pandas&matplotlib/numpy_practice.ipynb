## Difference Between Python `sum()` and numpy `np.sum()`

```python
#Python sum() is slower to np.sum()
#use python sum() only for python dataypes 
#use np.sum() only for numpy datatype
```

```python
#Visual representation of comparison, %timeit is a special method in jupyter to tell the time taken to execute that cell
massive_array=np.random.random(10000)
massive_array[:10]
```

```python
#applying python Sum() with %timeit to check the execution time
%timeit sum(massive_array)
```

```python
#applying np.sum()
%timeit np.sum(massive_array)
```

```python
#calculating standard deviation 
#Standard Deviation is the Measure of how Spread out the Group of numbers is from the Mean
std=np.std(a2)
```

```python
#calculating variance
#Variance is the Measure of Average Degree by which the number differs from the mean 
#Higher Variance=Higher range of Numbers 
#Lower Variance=Lower range of Numbers
var=np.var(a2)
```

```python
#Standard deviation =sqrt(var)
std==np.sqrt(var)
```

## Dot Product Vs Element Wise Product

```python
#Creating some random numbers in matrices
np.random.seed(seed=0)
mat1=np.random.randint(10,size=(3,4))
mat2=np.random.randint(10,size=(3,4))
mat1.shape,mat2.shape
```

```python
mat1
```

```python
mat2
```

```python
#Element Wise Product (Hadamard Multiplication)
mat1*mat2
```

```python
#Dot Product: the dim. of Column of Matrice 1 should be = to dim. of Row in Matrice 2
#but here both the matrices have dim (3,4)
#To make it work we need to tranpose one matrice (3,4).(4,3)=(3,3)
mat3=np.dot(mat1,mat2.T)
mat3,mat3.shape
```

## Exercise: Nut Butter Store Sales

```python
#Creating the data from scratch
np.random.seed(0)
store_sales=np.random.randint(10,size=(5,3))
store_sales
```

```python
#Creating a dataframe index=5 days of the week, labels=types of butter
store_sales_dataframe=pd.DataFrame(store_sales,
                                  index=["Mon","Tues","Wed","Thurs","Fri"],
                                  columns=["Almond Butter","Peanut Butter","Cashew Butter"])
store_sales_dataframe
```

```python
# Creating price per jar for the types of butter
np.random.seed(0)
jar_price=np.random.randint(10,14,size=(1,3))
jar_price
```

```python
#Creating Dataframe for this jar_price
jar_price_dataframe=pd.DataFrame(jar_price,
                                index=["Price"],
                                columns=["Almond Butter","Peanut Butter","Cashew Butter"])
```

```python
jar_price_dataframe
```

```python
#Calculating total amount sold per day using dot product and storing it in Total ($) column
store_sales_dataframe["Total ($)"]=np.dot(store_sales_dataframe,jar_price_dataframe.T)
```
