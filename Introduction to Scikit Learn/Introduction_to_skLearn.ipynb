# Introduction to Scikit-Learn

## 0. An End to End Scikit-Learn Workflow 

```python
# Importing data 
import pandas as pd
heart_disease=pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/refs/heads/master/data/heart-disease.csv")
heart_disease.head()
```

```python
# Creating Features Matrix
X=heart_disease.drop("target",axis=1)
# Creating Label
Y=heart_disease["target"]
```

```python
# Choosing a model and hyperparameters
# Hyperparameters are like dials of the model with which we can do model tunining to make it better or worse :D
from sklearn.ensemble import RandomForestClassifier 
clf=RandomForestClassifier()
#keeping default hyperparameter
clf.get_params()
```

```python
# Fit the Model to the Training Data
# For that we need to split the data into train and test

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
# Now Lets Fit the model to train data
clf.fit(x_train,y_train)
```

```python
# Now lets Predict some output on test data
y_preds=clf.predict(x_test)
y_preds
```

```python
y_test
```

```python
# Time for Evaluation on both train data and test data
clf.score(x_train,y_train)
```

```python
# Ahhh! Maximum accuracy on training data set 
# lets Score it on test data 
clf.score(x_test,y_test)
```

```python
# 78% accuracy ok not bad
```

```python
from sklearn.metrics import classification_report, confusion_matrix , accuracy_score
```

```python
print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

Three steps we have to follow:

    1. Splitting the Data into Features and Labels `X` and `Y`
    2. Filling (Also called imputing) and disregarding the missing values
    3. Converting Non-Numerical Values to Numerical data (this is known as Feature Encoding)

```python
# lets view the data 
heart_disease.head()
```

```python
# Lets Split it into Features X and Labels Y
X=heart_disease.drop("target",axis=1)
Y=heart_disease["target"]
```

```python
#Split the data into train test 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
# As the heart disease data does not have any missing value we are going to import another dataset
car_sales=pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/refs/heads/master/data/car-sales-extended.csv")
car_sales.head()
```

```python
car_sales.dtypes
```

```python
# Here we are going to split the data into labels X (Features) and Y (Label=Price)
X=car_sales.drop("Price",axis=1)
Y=car_sales["Price"]
```

```python
# Now we are going to split the data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
# Importing a model
from sklearn.ensemble import RandomForestRegressor # earlier we used RandomForestClassifier this time Regressor why ? cz we are going to predict some numeric data

model=RandomForestRegressor() #Initialising the Model
model.fit(x_train,y_train)  #fitting the model
model.score(x_test,y_test)  #scoring the model
```

```python
# AAHH! what is this error now. OOH we need to convert the string data into numeric data 
# Let's see how scikit learn help us to do that
```

```python
# Turn the categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features=["Make","Colour","Doors"]
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("one_hot",one_hot,categorical_features)],
                             remainder="passthrough")
transformed_X=transformer.fit_transform(X)
```

```python
#What is One Hot Encoder ? -> it converts the category into columns and put zero and one respectively
#E.g: for a car it has colour red
#there will be other colours too, White, Black,Red
# to represent the colour <Column Name=0 or 1>, White=0, Black=0 and Red=1
```

### 1.2 What if there were Missing Data?

1. We can fill them with some Values also called Imputation. Cons: Those filled data are not real
2. We can remove the missing data. Cons: Decrease in size of Data

```python
#importing missing car sales data
car_sales=pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/refs/heads/master/data/car-sales-extended-missing-data.csv")
car_sales.head()
```

```python
# In a huge data set how to know if there are any missing data?
# we can use this :
car_sales.isna().sum()
```

```python
# Here it shows the number of missing data for each columns
```

```python
#Creating X and Y
X=car_sales.drop("Price",axis=1)
Y=car_sales["Price"]
```

### Option 1: Filling Missing Data using Pandas

```python
# Fill the "Make" Column 
car_sales["Make"]=car_sales["Make"].fillna("missing")

# Fill the "Colour" Column
car_sales["Colour"]=car_sales["Colour"].fillna("missing")

# Fill the "Odometer (KM)" Column
car_sales["Odometer (KM)"]=car_sales["Odometer (KM)"].fillna(car_sales["Odometer (KM)"].mean())

#Fill the "Doors" Column
car_sales["Doors"]=car_sales["Doors"].fillna(4)
```

```python
car_sales.isna().sum()
```

```python
# So we have Filled the Missing data for the Categories Column, Then what about the Price Column
# As we are going to Predict the price column we are not going to fill the data with some non real value , we are going to Drop it
car_sales=car_sales.dropna()
```

```python
car_sales.isna().sum() #Finally no more Missing Data, Great Job!!
```

```python
#Creating X and Y
X=car_sales.drop("Price",axis=1)
Y=car_sales["Price"]
```

```python
# Turn the categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features=["Make","Colour","Doors"]
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("one_hot",one_hot,categorical_features)],
                             remainder="passthrough")
transformed_X=transformer.fit_transform(X)
```
