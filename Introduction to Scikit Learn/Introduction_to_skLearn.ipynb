### 1.2 What if there were Missing Data?

```python
#importing missing car sales data
car_sales=pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/refs/heads/master/data/car-sales-extended-missing-data.csv")
car_sales.head()
```

### Option 1: Filling Missing Data using Pandas

```python
# Fill the "Make" Column 
car_sales["Make"]=car_sales["Make"].fillna("missing")

# Fill the "Colour" Column
car_sales["Colour"]=car_sales["Colour"].fillna("missing")

# Fill the "Odometer (KM)" Column
car_sales["Odometer (KM)"]=car_sales["Odometer (KM)"].fillna(car_sales["Odometer (KM)"].mean())

#Fill the "Doors" Column
car_sales["Doors"]=car_sales["Doors"].fillna(4)
```

```python
car_sales.isna().sum()
```

```python
# So we have Filled the Missing data for the Categories Column, Then what about the Price Column
# As we are going to Predict the price column we are not going to fill the data with some non real value , we are going to Drop it
car_sales=car_sales.dropna()
```

```python
car_sales.isna().sum() #Finally no more Missing Data, Great Job!!
```

```python
#Creating X and Y
X=car_sales.drop("Price",axis=1)
Y=car_sales["Price"]
```

```python
# Turn the categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features=["Make","Colour","Doors"]
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("one_hot",one_hot,categorical_features)],
                             remainder="passthrough")
transformed_X=transformer.fit_transform(X)
```

```python
transformed_X
```

### Option 2: Filling Data using Scikit Learn

```python
#Reimporting the Dataset 
car_sales_missing=pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/refs/heads/master/data/car-sales-extended-missing-data.csv")
car_sales_missing.head()
```

### Best Practice is to handle the missing data after splitting the data into train and test 

#### Here we are just going to see the process of handling missing data(Not Recommended in real life workflow)

```python
# Drop the rows with no Price Labels
car_sales_missing.dropna(subset=["Price"],inplace=True)
car_sales_missing.isna().sum()
```

```python
# Splitting into X and Y
X=car_sales_missing.drop("Price",axis=1)
Y=car_sales_missing["Price"]
```

```python
# Filling Missing Value using Scikit Learn
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

#Filling Categorical values with "missing" and numerical values with mean
cat_imputer=SimpleImputer(strategy="constant",fill_value="missing")
door_imputer=SimpleImputer(strategy="constant",fill_value=4)
num_imputer=SimpleImputer(strategy="mean")

# Defining Columns
cat_features=["Make","Colour"]
door_feature=["Doors"]
num_features=["Odometer (KM)"]

# Create an imputer 
imputer=ColumnTransformer([
    ("cat_imputer",cat_imputer,cat_features),
    ("door_imputer",door_imputer,door_feature),
    ("num_imputer",num_imputer,num_features)
])
```

```python
#checking the number of missing values in X
X.isna().sum()
```

```python
# Lets fill now 
filled_X=imputer.fit_transform(X)
filled_X
```

```python
# Lets conver the array to a datafram 
car_sales_filled=pd.DataFrame(filled_X,columns=["Make","Colour","Doors","Odometer (KM)"])
car_sales_filled.head()
```

```python
# Is there any missing data Now ?
car_sales_filled.isna().sum()
```

```python
# As we have handled the Missing Values lets convert the non-numerical data to numerical usng OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features=["Make","Colour","Doors"]
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("one_hot",one_hot,categorical_features)],
                             remainder="passthrough")
transformed_X=transformer.fit_transform(car_sales_filled)
transformed_X
```

```python
# now lets initialise the model and fit it 
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(transformed_X,Y,test_size=0.2)
model=RandomForestRegressor()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

**Area Under the Receiver Operating Characteristic Curve (AUC/ROC)**

* Area Under Curve (AUC)
* ROC Curve

ROC Curve is a comparision of a Model's True Positive Rate (TPR) VS False Positive Rate (FPR)

What are these True Positive , False Positive?

* True Positive: When the Model predicts 1 and the actual value is 1 (Model: Has Heart Disease , Actual: Has Heart Disease)
* False Positive: When the Model predicts 1 and the actual value is 0 (Model: Has Heart Disease, Actual: Doesn't have Heart Disease)
* True Negative: When the Model predicts 0 and the actual value is 0  (Model: Doesn't Have Heart Disease, Actual: Doesn't Have Heart Disease)
* False Negative: When the Model predicts 0 and the actual value is 1 (Model: Doesn't Have Heart Disease, Actual: Has Heart Disease)

```python
# Predicting the probabilities 
y_probs=clf.predict_proba(x_test)
y_probs[:10]
```

```python
# [0,1]=[Doesn't have Heart Disease, Have Heart Disease]=[False,True]
# Splitting by true values
y_positive_probs=y_probs[:,1]
y_positive_probs[:10]
```

```python
# calculating FPR , TPR and Thresholds
from sklearn.metrics import roc_curve

fpr,tpr,thresholds=roc_curve(y_test,y_positive_probs)

# Check fpr
fpr[:10]
```

```python
#Lets Visualise it 
import matplotlib.pyplot as plt

def plot_roc_curve(fpr,tpr):
    """
    Plot a ROC Curve based on False Positive rate and True Positive Rate
    """

    plt.plot(fpr,tpr,color="orange",label="ROC")
    plt.plot([0,1],[0,1],label="Guessing", color="darkblue", linestyle="--")

    # Customizing the plot
    plt.xlabel("False Positive Rate FPR")
    plt.ylabel("True Positive Rate TPR")
    plt.title("Receiver Operating Characteristic Curve ROC Curve")
    plt.legend()
    plt.show()

plot_roc_curve(fpr,tpr)
```

```python
# Calculating AUC Score 
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test,y_positive_probs)
```

```python
# AUC is defined as the area under the ROC curve. AUC tells us how well a model is classifying .
# ROC Curve and AUC score is used only for Binary Classification
```

```python
def clf_evaluation(y_true,y_pred):
    """
    Performs Evaluation Comparision for Classification model Y_True Labels Vs Y_Pred Labels
    """
    accuracy=accuracy_score(y_true,y_pred)
    precision=precision_score(y_true,y_pred)
    recall=recall_score(y_true,y_pred)
    f1=f1_score(y_true,y_pred)

    #Creating a dict for proper display
    metric_dict={"Accuracy Score:":np.round(accuracy,2),
                "Precision Score:":np.round(precision,2),
                "Recall Score:":np.round(recall,2),
                "F1 Score:":np.round(f1,2)}
    return metric_dict
    
```
